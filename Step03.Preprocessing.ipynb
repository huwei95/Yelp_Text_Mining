### Remove of Stopwords and punctuation

```python
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer(r'\w+') #use this filter out puctuation.

Remove_stopw_punc=[]
for item in doc_lower:
    subitems_Filtered=[]
    subitems_After_Filter=[]
    for subitem in tokenizer.tokenize(item):
        if (subitem in punctuation_list) or (subitem in stopword_list):
            subitems_Filtered.append(subitem)
        else:
            subitems_After_Filter.append(subitem)
    Remove_stopw_punc.append(' '.join(subitems_After_Filter))
```

### test of lemmatizing

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer

lemma = WordNetLemmatizer()
stemmer=PorterStemmer()

for item in Remove_stopw_punc:
    subitems_list=item.split(' ')
    for subitem in subitems_list: 
        item_lemmatized=lemma.lemmatize(subitem)
        item_stemmered=stemmer.stem(subitem)
        if (item_lemmatized==subitem) & (item_lemmatized<>item_stemmered) :
            print subitem, '-> (lemmatize)', item_lemmatized
            print subitem, '-> (stem)', item_stemmered
            print '-------------------------------------'
```

### 3.2 Implementation of Preprocessing

```python
start_time=time.time()
Preprocessed_Text=[]
tokenizer=RegexpTokenizer(r'\w+') #use this filter out puctuation.
lemma = WordNetLemmatizer()
stemmer=PorterStemmer()

for i in range(len(Yelp_us_food_review)):
    if i%10000==0:
        small_cycle_start_time=time.time()
    doc_lower=Yelp_us_food_review.iloc[i]['text'].lower().decode("utf8")
    subitems_After_Filter=[]
    for subitem in tokenizer.tokenize(doc_lower):
        if (subitem not in punctuation_list) & (subitem not in SEO_Stop_Words):
            subitems_lemmatized=lemma.lemmatize(subitem)
            subitems_After_Filter.append(subitems_lemmatized)
#             if subitem<>subitems_lemmatized:
#                 print subitem,'->', subitems_lemmatized
    Preprocessed_Text.append(' '.join(subitems_After_Filter))
    if (i%10000==9999) or (i== len(Yelp_us_food_review)) :
        small_cycle_end_time=time.time()
        print (i+1)/10000, ' 10k processed: ', (small_cycle_end_time-small_cycle_start_time)/60, 'mins'

Yelp_us_food_review['Preprocessed_Text']=Preprocessed_Text
Yelp_us_food_review.to_csv('Yelp_us_food_review_preprocessed.csv',index=False,encoding='utf-8')
end_time=time.time()
print "Total %s mins" % ((end_time-start_time)/60)
```
